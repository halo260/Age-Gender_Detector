Blurring (Smoothing)

```python
img3 = cv2.imread("Resources\Har Har Sambhu.jpeg")
plt.imshow(img3)
```

```python
img  = cv2.cvtColor(img3, cv2.COLOR_BGR2HSV)#converts image from BGR color space to HSV(Hue, Saturation, Value)
figure_size = 2
result = cv2.blur(img, (figure_size, figure_size))
plt.figure(figsize = (11,6))
plt.subplot(121), plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2HSV)), plt.title("Original")
plt.xticks([]), plt.yticks([])
plt.subplot(131), plt.imshow(cv2.cvtColor(result,cv2.COLOR_BGR2HSV)), plt.title("Blurred")
plt.xticks([]), plt.yticks([])
plt.show
```

Resizing

```python
img4 = cv2.imread("Resources\cacf84ee-c123-4dac-bd41-47dc71fe386f.jpeg")
(height, width) = img4.shape[0:2] #store variable in tuple
zoom = cv2.resize(img4, None, fx=4, fy=4, interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
plt.title("Shiv")
```

```python
zoom = cv2.resize(img4, (int(width*2), int(height*2)), interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

Rotation 

```python
m = cv2.getRotationMatrix2D((width/2, height/2), 30, 1)
rotate = cv2.warpAffine(img3, m, (height, width))
plt.imshow(rotate)
```
